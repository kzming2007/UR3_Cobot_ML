```python

import seaborn as sns

temp_cols = ['Temperature_T0', 'Temperature_J1', 'Temperature_J2', 'Temperature_J3', 'Temperature_J4', 'Temperature_J5']
curr_cols = ['Current_J0', 'Current_J1', 'Current_J2', 'Current_J3', 'Current_J4', 'Current_J5']

# 6개 x 6개 조합이므로 총 36개의 칸(6행 6열)을 생성합니다.
fig, axes = plt.subplots(6, 6, figsize=(20, 20))

# 전체 격자의 인덱스를 추적하기 위한 카운터 변수
ax_idx = 0

for col in temp_cols:
    for curr in curr_cols:  # 문자열 컬럼명만 순서대로 가져옴
        # 2차원 배열 구조의 axes에서 정확한 위치 지정
        ax = axes.flatten()[ax_idx]

        # 그래프 생성
        sns.scatterplot(x=curr, y=col, data=df, ax=ax, alpha=0.3, s=10)
        ax.set_title(f'{col} vs {curr}', fontsize=9)

        # 다음 칸으로 이동
        ax_idx += 1

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
import pandas as pd

# [해결 핵심] 결측치를 제거(또는 False 채움)하고 명확하게 bool 타입으로 변환
df_clean = df[['Robot_ProtectiveStop', 'grip_lost']].dropna().copy()
df_clean['Robot_ProtectiveStop'] = df_clean['Robot_ProtectiveStop'].astype(bool)
df_clean['grip_lost'] = df_clean['grip_lost'].astype(bool)

# 1. 혼동 행렬 데이터 계산 (정제된 데이터프레임 사용)
cm = confusion_matrix(y_true=df_clean['Robot_ProtectiveStop'],
                      y_pred=df_clean['grip_lost'],
                      labels=[True, False])

# 2. 히트맵 시각화
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['True', 'False'],
            yticklabels=['True', 'False'])

plt.title('Confusion Matrix (Robot Stop vs Grip Lost)')
plt.ylabel('Actual (Robot_ProtectiveStop)')
plt.xlabel('Predicted (grip_lost)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 로봇이 실제로 정지한 위험 순간(True) 데이터만 필터링
df_stop = df[df['Robot_ProtectiveStop'] == True]

# 데이터가 존재하는지 확인용 출력
print(f"전체 데이터 중 로봇 정지 순간의 데이터 개수: {len(df_stop)}건")

temp_cols = ['Temperature_T0', 'Temperature_J1', 'Temperature_J2', 'Temperature_J3', 'Temperature_J4', 'Temperature_J5']
curr_cols = ['Current_J0', 'Current_J1', 'Current_J2', 'Current_J3', 'Current_J4', 'Current_J5']

# 6행 6열 구조의 서브플롯 생성
fig, axes = plt.subplots(6, 6, figsize=(20, 20))
ax_idx = 0

for col in temp_cols:
    for curr in curr_cols:
        ax = axes.flatten()[ax_idx]

        # 필터링된 데이터(df_stop)를 사용해 산점도 시각화
        # 데이터 수가 줄어들었으므로 굳이 샘플링하지 않고 전체를 표현합니다.
        sns.scatterplot(x=curr, y=col, data=df_stop, ax=ax, alpha=0.6, s=15, color='crimson')
        ax.set_title(f'{col} vs {curr} (Stop)', fontsize=9)

        ax_idx += 1

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 로봇 정지 순간(True) 데이터 필터링 및 카피
df_stop = df[df['Robot_ProtectiveStop'] == True].copy()

temp_cols = ['Temperature_T0', 'Temperature_J1', 'Temperature_J2', 'Temperature_J3', 'Temperature_J4', 'Temperature_J5']
curr_cols = ['Current_J0', 'Current_J1', 'Current_J2', 'Current_J3', 'Current_J4', 'Current_J5']

# 2. 모든 전류 컬럼에 대해 절대값 파생 변수 생성 및 리스트 갱신
abs_curr_cols = []
for curr in curr_cols:
    abs_col_name = f'Abs_{curr}'
    df_stop[abs_col_name] = df_stop[curr].abs()
    abs_curr_cols.append(abs_col_name)

# 3. 6행 6열 구조의 서브플롯 생성
fig, axes = plt.subplots(6, 6, figsize=(20, 20))
ax_idx = 0

for col in temp_cols:
    for abs_curr in abs_curr_cols:
        ax = axes.flatten()[ax_idx]

        # x축에 절대값 처리된 전류 변수 적용
        sns.scatterplot(x=abs_curr, y=col, data=df_stop, ax=ax, alpha=0.6, s=15, color='darkorange')
        ax.set_title(f'{col} vs {abs_curr} (Stop)', fontsize=9)

        ax_idx += 1

plt.tight_layout()
plt.show()
```
